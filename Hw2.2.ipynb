```python
# Частка пропусків у кожній колонці
missing_data_percent = df.isnull().sum() / len(df) * 100
print(missing_data_percent)

import matplotlib.pyplot as plt
columns_with_missing_data = missing_data_percent[missing_data_percent > 0]
plt.figure(figsize=(10, 6))
columns_with_missing_data.plot(kind='bar', color='purple')
plt.title('Частка пропусків у кожній колонці')
plt.xlabel('Колонка')
plt.ylabel('Частка пропусків (%)')
plt.show()
```

```python
# Зберігаємо стовпець "Мова програмування"
language_column = df["Мова програмування"]

# Видалити всі стовпці з пропусками
df_cleaned = df.dropna(axis=1)

# Додаємо стовпець "Мова програмування" назад, використовуючи pd.concat
df_cleaned = pd.concat([df_cleaned, language_column], axis=1)
```

```python
# Перевірка частки пропусків після видалення
df_cleaned.isnull().sum() / len(df_cleaned) * 100
```

```python
# Видалення всіх рядків з пропусками
df_cleaned = df_cleaned.dropna()
```

```python
# Новий розмір таблиці
df_cleaned.shape
```

```python
# Таблиця зі спеціалістами, які використовують Python
python_data = df_cleaned[df_cleaned["Мова програмування"].str.contains("Python", na=False)]
print(python_data)
```

```python
# Розмір таблиці python_data
python_data.shape
```

```python
# Групування за стовпцем "Посада"
grouped_data = python_data.groupby('Посада')
```

```python
# Агрегування даних для мінімальних та максимальних значень зарплати
agg_salary_data = grouped_data.agg({'Зарплата на місяць': ['min', 'max']})
agg_salary_data
print(agg_salary_data)

plt.figure(figsize=(10, 6))
for name, group in grouped_data:
    plt.scatter(group['Посада'], group['Зарплата на місяць'], label=name)
plt.title('Діаграма розсіювання зарплат за місяць для всіх посад')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.legend(title='Посада', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Функція для обчислення середньої зарплати
def fill_avg_salary(row):
    return (row[('Зарплата на місяць', 'min')] + row[('Зарплата на місяць', 'max')]) / 2

# Додавання нового стовпця із середньою зарплатою
agg_salary_data['avg'] = agg_salary_data.apply(fill_avg_salary, axis=1)

print(agg_salary_data)
```

```python
import seaborn as sns

# Завантаження даних
df = pd.read_csv('C:/Users/MikeK/GoIT/2017_jun_final - 2017_jun_final.csv')

def convert_experience_to_numeric(experience):
    """Функція для конвертації досвіду роботи з текстового в числовий формат."""
    if pd.isna(experience):
        return None
    if 'менше 3 місяців' in experience:
        return 0.25  # Переводимо "менше 3 місяців" у 0.25 року
    if 'більше 10 років' in experience:
        return 10  # Переводимо "більше 10 років" у 10 років
    try:
        years = float(experience.split()[0])
        return years
    except ValueError:
        return None

def analyze_salary_dependency_line_plot(df):
    # Вибір потрібних стовпців для аналізу
    columns_of_interest = ['Місто', 'Розмір компанії', 'Загальний досвід роботи', 
                           'Досвід роботи на поточному місці', 'Мова програмування', 'Посада', 'Зарплата на місяць']

    # Фільтрація даних
    df_filtered = df[columns_of_interest].copy()

    # Перетворення досвіду роботи в числовий формат
    df_filtered['Загальний досвід роботи'] = df_filtered['Загальний досвід роботи'].apply(convert_experience_to_numeric)
    df_filtered['Досвід роботи на поточному місці'] = df_filtered['Досвід роботи на поточному місці'].apply(convert_experience_to_numeric)

    # Перетворення міст у числовий формат за допомогою One-Hot Encoding
    df_encoded = pd.get_dummies(df_filtered, columns=['Місто'], drop_first=True)

    # Лінійна діаграма для кореляції між загальним досвідом роботи та зарплатою
    plt.figure(figsize=(12, 8))
    sns.regplot(x='Загальний досвід роботи', y='Зарплата на місяць', data=df_filtered, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
    plt.title('Кореляція між загальним досвідом роботи та зарплатою')
    plt.xlabel('Загальний досвід роботи (роки)')
    plt.ylabel('Зарплата на місяць')
    plt.show()

    # Лінійна діаграма для кореляції між досвідом роботи на поточному місці та зарплатою
    plt.figure(figsize=(12, 8))
    sns.regplot(x='Досвід роботи на поточному місці', y='Зарплата на місяць', data=df_filtered, scatter_kws={'alpha':0.5}, line_kws={'color': 'blue'})
    plt.title('Кореляція між досвідом роботи на поточному місці та зарплатою')
    plt.xlabel('Досвід роботи на поточному місці (роки)')
    plt.ylabel('Зарплата на місяць')
    plt.show()

    # Лінійна діаграма для кореляції між містом та зарплатою
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Місто', y='Зарплата на місяць', data=df_filtered, ci=None, palette='viridis')
    plt.xticks(rotation=90)
    plt.title('Кореляція між містом та зарплатою')
    plt.xlabel('Місто')
    plt.ylabel('Зарплата на місяць')
    plt.show()



# Виклик функції для аналізу
analyze_salary_dependency_line_plot(df)
```
